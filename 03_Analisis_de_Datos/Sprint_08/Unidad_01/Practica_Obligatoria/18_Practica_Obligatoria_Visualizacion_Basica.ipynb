2. Histograma del área total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamaño 18, color azul acero)

```python

#plt.hist(df["area_total_km2"], bins=50, color="steelblue")
#plt.legend();


fig,ax = plt.subplots()
ax.hist(df["area_total_km2"], bins=50, color="steelblue", label = "Hist Area")
font_dict = {
    "size": 18,
    "family": "Bradley Hand itc",
    "color": "black"
}
ax.set_xlabel("Superficie ($Km^2~$)", fontdict=font_dict)
ax.set_ylabel("Númenro de Ciudades", fontdict=font_dict)
ax.legend();
```

3. El scatter de área de agua de un ciudad con su poblacion.

```python
x3 = df["area_water_km2"]
y3 = df["population_total"]/1000
plt.scatter(x3,y3, color="green", alpha=0.5 )
plt.xlabel("Area de agua en la ciudad ($Km^2~$)")
plt.ylabel("Población (miles de hanbitantes)")
plt.suptitle("Población por $Km^2~$ de agua")

max_pop = df["population_total"].max()
x_max_pop = df.loc[df["population_total"] == max_pop].area_water_km2.to_list()[0]

max_agu = df["area_water_km2"].max()
y_max_agu = df.loc[df["area_water_km2"] == max_agu].population_total.to_list()[0]

coordenadas = (x_max_pop,max_pop/1000)

props_flecha = {
    "arrowstyle": "->",
    "connectionstyle":"arc3,rad=0"
    }
plt.annotate("Outliers", xy =coordenadas, textcoords='offset points', xytext=(-80, -50),
             arrowprops=props_flecha)

coordenadas = (max_agu,y_max_agu/1000)
plt.annotate("", xy =coordenadas, textcoords='offset points', xytext=(-140, 100),
             arrowprops=props_flecha)
```

4. El boxplot de la altura de las ciudades.

```python
plt.boxplot(df.elevation_m, vert=False, whis=3, patch_artist = True)
plt.title("Diagrama de caja de la Altitud de las cuidades de California")
plt.xlabel("Altitud (m)");
```
