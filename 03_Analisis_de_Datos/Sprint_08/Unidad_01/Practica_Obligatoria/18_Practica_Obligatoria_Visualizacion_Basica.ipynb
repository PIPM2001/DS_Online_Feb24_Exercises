### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)
2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamao 18, color azul acero)
3. El scatter de rea de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
ind_5 = df["population_total"].sort_values(ascending=False).head(5).to_list()
df_5max = df.query(f"population_total in {ind_5}")
df_5max
```

```python
df_5max["pop_k"] = round(df_5max["population_total"]/1000,3)
```

```python
x1 = df_5max.sort_values("pop_k", ascending=False)["city"]
y1 = df_5max.sort_values("pop_k", ascending=False)["pop_k"]

fig,ax = plt.subplots(1,1, figsize =(10,10))
ax.bar(x1,y1, color = ["steelblue","grey", "red","green", "brown"], alpha=.5, width=.7)
ax.set_title("Ciudades mas pobladas de California")
ax.set_xlabel("Ciudades")
ax.set_ylabel("Habitantes (miles)");

for p in ax.patches:
                height = p.get_height()
                ax.annotate(f'{height:.2f}', (p.get_x() + p.get_width() / 2., height), 
                            ha='center', va='center', xytext=(0, 9), textcoords='offset points')
```

2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamao 18, color azul acero)

```python
fig,ax = plt.subplots()
ax.hist(df["area_total_km2"], bins=50, color="steelblue", label = "Hist Area")
font_dict = {
    "size": 18,
    "family": "Bradley Hand itc",
    "color": "black"
}
ax.set_xlabel("Superficie ($Km^2~$)", fontdict=font_dict)
ax.set_ylabel("Nmenro de Ciudades", fontdict=font_dict)
ax.legend();
```

3. El scatter de rea de agua de un ciudad con su poblacion.

```python
x3 = df["area_water_km2"]
y3 = df["population_total"]/1000
plt.scatter(x3,y3, color="green", alpha=0.5 )
plt.xlabel("Area de agua en la ciudad ($Km^2~$)")
plt.ylabel("Poblacin (miles de hanbitantes)")
plt.suptitle("Poblacin por $Km^2~$ de agua")

max_pop = df["population_total"].max()
x_max_pop = df.loc[df["population_total"] == max_pop].area_water_km2.to_list()[0]

max_agu = df["area_water_km2"].max()
y_max_agu = df.loc[df["area_water_km2"] == max_agu].population_total.to_list()[0]

coordenadas = (x_max_pop,max_pop/1000)

props_flecha = {
    "arrowstyle": "->",
    "connectionstyle":"arc3,rad=0"
    }
plt.annotate("Outliers", xy =coordenadas, textcoords='offset points', xytext=(-80, -50),
             arrowprops=props_flecha)

coordenadas = (max_agu,y_max_agu/1000)
plt.annotate("", xy =coordenadas, textcoords='offset points', xytext=(-140, 100),
             arrowprops=props_flecha)
```

4. El boxplot de la altura de las ciudades.

```python
plt.boxplot(df.elevation_m, vert=False, whis=3, patch_artist = True)
plt.title("Diagrama de caja de la Altitud de las cuidades de California")
plt.xlabel("Altitud (m)");
```
